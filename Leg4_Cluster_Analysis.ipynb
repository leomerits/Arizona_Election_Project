# Preprocessing the Data

```python
import pandas as pd
from pathlib import Path
```

```python
#file path will be update to the table created in our sql database
file_path = "/Users/joshzaragoza/Arizona_Election_Project/Leg_District_4_clean.csv"
df_voters = pd.read_csv(file_path)
df_voters.head(10)
```

```python
#Drop all columns contained unecesarry features or null nan 
df_voters.drop(columns=["Voter ID","CodedRaceName","LastName","FirstName","MiddleName","Suffix","HD","CivSpan","2020:VolProp"], inplace=True)
df_voters
```

```python
# Columns
df_voters.count()
```

```python
#seeing all data types
df_voters.dtypes
```

```python
df_voters.isnull()
```

```python
df_voters.isnull().sum()
```

```python
#drop any null or nan in the data frame 
df_voters.dropna(axis = 0, how = "any", thresh = None, subset = None, inplace=True)
df_voters
```

```python
df_voters.isnull().sum()
```

```python
#Transform String Columns
def change_string(Ethnicity):
    if Ethnicity == "White":
        return 1
    if Ethnicity == "Hispanic": 
        return 2
    else:
        return 3
    
df_voters["Ethnicity"] = df_voters["Ethnicity"].apply(change_string)
df_voters.head()
```

```python
#Transform String Columns
def change_string(Sex):
    if Sex == "M":
        return 1
    else: 
        return 0
    
df_voters["Sex"] = df_voters["Sex"].apply(change_string)
df_voters.head()
```

```python
#Transform String Columns
def change_string(PartyName):
    if PartyName == "Democrat":
        return 1
    if PartyName == "Republican": 
        return 2
    if PartyName == "Other":
        return 3
    
df_voters["PartyName"] = df_voters["PartyName"].apply(change_string)
df_voters.head()
```

```python
df_voters['PartisanScore'].astype(int)
df_voters
```

```python
df_voters = df_voters.astype({'PartisanScore':'int', 'NatTO':'int'})
df_voters
```

```python
df_voters = df_voters.astype({'Partisanship':'int','2020:CivKidsinHH':'int','2020:CivLibIdeo':'int','2020:CivMarriage':'int'})
df_voters
```

```python
def convert_score(PartisanScore):
    if PartisanScore >= 35 and PartisanScore <= 65:
        return 1
    else:
        return 0
    
df_voters["Outcome"] = df_voters["PartisanScore"].apply(convert_score)
df_voters.head()
```

```python
df_voters
```

```python
# Saving cleaned data
file_path = "/Users/joshzaragoza/Arizona_Election_Project/Leg_District_4_cleaned.csv"
df_voters.to_csv(file_path, index=False)
```

```python
import pandas as pd
import plotly.express as px
import hvplot.pandas
from sklearn.cluster import KMeans
```

```python
#loading cleaned data
file_path = "/Users/joshzaragoza/Arizona_Election_Project/Leg_District_4_cleaned.csv"
df_voters = pd.read_csv(file_path)
df_voters.head(10)
```

```python
#Drop all columns contained unecesarry features  
df_voters.drop(columns=["Zip"], inplace=True)
df_voters
```

```python
# Initializing model with K = 2
model = KMeans(n_clusters=3, random_state=5)
model
```

```python
# Fitting model
model.fit(df_voters)
```

```python
# Get the predictions
predictions = model.predict(df_voters)
print(predictions)
```
