```python
# Import our dependencies
import pandas as pd
import matplotlib as plt
import sklearn as skl
import tensorflow as tf
from path import Path
import numpy as np
```

```python
#file path will be update to the table created in our sql database
file_path = "/Users/joshzaragoza/Arizona_Election_Project/Resources/Voter Data/Leg_District4.csv"
df_voters = pd.read_csv(file_path)
df_voters.head(10)
```

```python
df_voters.drop(columns=["Zip","LastName","FirstName","MiddleName"], inplace=True)
df_voters
```

```python
# Columns
df_voters.count()
```

```python
df_voters.isnull()
```

```python
#drop any null or nan in the data frame 
df_voters.dropna(axis = 0, how = "any", thresh = None, subset = None, inplace=True)
df_voters
```

```python
df_voters.isnull().sum()
```

```python
df_voters.dtypes
```

```python
# Convert Float64 to int dtypes
df_voters = df_voters.astype({'Voter Score':'int','Turnout Score':'int','Partisanship Score':'int','Kids in HH':'int','Liberal Ideology':'int'})
df_voters
```

```python
# Rename column 1
df_voters.rename(columns={'PartyName':'Party'}, inplace=True)
df_voters.head()
```

```python
# Create a list of our conditions
conditions = [
    (df_voters['Voter Score'] <= 34),
    (df_voters['Voter Score'] >= 35) & (df_voters['Voter Score'] <= 65),
    (df_voters['Voter Score'] >= 66) & (df_voters['Voter Score'] <=100)
    ]

# Create of values we want assigned to the conditions
values = ['False', 'True', 'False']

# Create a new column with np.select to assign values to it using our lists as arguments
df_voters['Swing Voter'] = np.select(conditions, values)

# Display updated DataFrame
df_voters
```

```python
# Create a list of our conditions
conditions = [
    (df_voters['Age'] >= 18) & (df_voters['Age'] <= 24),
    (df_voters['Age'] >= 25) & (df_voters['Age'] <= 34),
    (df_voters['Age'] >= 35) & (df_voters['Age'] <=44),
    (df_voters['Age'] >= 45) & (df_voters['Age'] <=54),
    (df_voters['Age'] >= 55) & (df_voters['Age'] <=64),
    (df_voters['Age'] >= 65),
    ]

# Create of values we want assigned to the conditions
values = ['1', '2', '3','4','5','6']

# Create a new column with np.select to assign values to it using our lists as arguments
df_voters['Age'] = np.select(conditions, values)

# Display updated DataFrame
df_voters
```

```python
# Changing to numerical by label encoder
from sklearn.preprocessing import LabelEncoder 
number = LabelEncoder()
df_voters["Party"] = number.fit_transform(df_voters["Party"])
df_voters["Sex"] = number.fit_transform(df_voters["Sex"])
df_voters["Ethnicity"] = number.fit_transform(df_voters["Ethnicity"])
df_voters
```

```python
#Drop columns that doesn't impact Swing Voter chances
df_voters.drop(columns=["Voter Score","Turnout Score","Partisanship Score","Kids in HH", "Liberal Ideology"], inplace=True)
df_voters
```

# Support Vectors Machine

```python
y = df_voters["Swing Voter"]
X = df_voters.drop(columns="Swing Voter")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
   y,  random_state=1, stratify=y)
X_train.shape
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
results = pd.DataFrame({
   "Prediction": y_pred,
   "Actual": y_test
}).reset_index(drop=True)
results.head()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

# Undersampling Majority 0 to Minority 1

```python
df = df_voters.copy()
```

```python
df.sample()
```

```python
df_zeros = df[df.SwingVoter == 0]
df_ones = df[df.SwingVoter == 1]
```

```python
df_zeros.shape
```

```python
df_zeros = df_zeros.sample(df_ones.shape[0])
```

```python
df_zeros.shape
```

```python
df_ones.shape
```

```python
undersample_df = pd.concat([df_zeros, df_ones])
```

```python
undersample_df.shape
```

```python
undersample_df.head(3)
```

```python
X = undersample_df.drop(['SwingVoter'], axis=1)
```

```python
y = undersample_df['SwingVoter']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```
