```python
#file path will be update to the table created in our sql database
file_path = "/Users/joshzaragoza/Arizona_Election_Project/AZ_Dummy_Voter_File.csv"
Arizona_Election_df = pd.read_csv(file_path)
Arizona_Election_df.head(10)
```

```python
#Drop all columns contained unecesarry features or null nan 
Arizona_Election_df.drop(columns=["Voter File VANID","mAddress","mCity","mState","mZip5","mZip4","Address","City","State","Zip5","Zip4","LastName","FirstName","MiddleName","Suffix","CD","SD","HD","PreferredEmail","Age","Turnout Score","Ethnicity"], inplace=True)
Arizona_Election_df
```

```python
#drop any null or nan in the data frame 
Arizona_Election_df.dropna(axis = 0, how = "any", thresh = None, subset = None, inplace=False)
Arizona_Election_df
```
